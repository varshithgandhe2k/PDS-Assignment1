```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_clean=pd.read_csv("../data_clean/clean_yield_data.csv")
```

```python
#Visualization - 1
plt.figure(figsize=(12, 6))
sns.boxplot(x='parental level of education', y='reading score', data=data_clean, palette='pastel')
plt.title('Parental Education Level vs. Reading Scores')
plt.xlabel('Parental Education Level')
plt.ylabel('Reading Score')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Visualization -2 : Distribution of Scores
plt.figure(figsize=(10, 6))
sns.histplot(data_clean['math score'], kde=True, color='blue', bins=30)
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization - 3 : Test Preparation Course Impact
plt.figure(figsize=(10, 6))
sns.barplot(x='test preparation course', y='math score', data=data_clean, ci=None, palette='Set2')
plt.title('Impact of Test Preparation Course on Average Scores')
plt.xlabel('Test Preparation Course')
plt.ylabel('Math Score')
plt.show()
```

```python
# Visualization - 4 : Scatter plot
plt.figure(figsize=(6, 4))
sns.scatterplot(x='math score', y='reading score', data=data_clean)
plt.title('Scatter Plot of math score vs. reading score')
plt.xlabel('math score')
plt.ylabel('reading score')
plt.show()
```

```python
# Visualization - 5 : Lineplot
plt.figure(figsize=(6, 4))
sns.lineplot(x='race/ethnicity', y='reading score', data=data_clean)
plt.title('Line plot of  Scores Over race/ethnicity')
plt.xlabel('race/ethnicity')
plt.ylabel('Reading Scores')
plt.show()
```

```python
# Visualization -6 : Bar chart
plt.figure(figsize=(6,4))
sns.countplot(x='parental level of education', data=data_clean)
plt.title('Bar Chart of parental level of education')
plt.xlabel('parental level of education')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
